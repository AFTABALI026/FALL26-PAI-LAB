# Water Jug Problem using Depth-First Search (DFS)

## Problem Definition
- **Jug 1 Capacity**: 3 liters
- **Jug 2 Capacity**: 5 liters
- **Goal**: Reach state (0, 4) - having 4 liters in Jug 2
- **Initial State**: (0, 0) - both jugs empty

## Available Operations:
1. Fill Jug 1
2. Fill Jug 2
3. Empty Jug 1
4. Empty Jug 2
5. Pour Jug 1 to Jug 2
6. Pour Jug 2 to Jug 1

```python
def water_jug_dfs(x, y, goal, visited, path):
    """DFS to solve water jug problem"""
    if (x, y) == goal:
        return True
    
    if (x, y) in visited:
        return False
    
    visited.add((x, y))
    
    # All possible moves
    moves = [
        ((3, y), "Fill Jug 1"),
        ((x, 5), "Fill Jug 2"),
        ((0, y), "Empty Jug 1"),
        ((x, 0), "Empty Jug 2"),
        ((max(0, x+y-5), min(5, x+y)), "Pour Jug 1→2"),
        ((min(3, x+y), max(0, x+y-3)), "Pour Jug 2→1")
    ]
    
    for (nx, ny), rule in moves:
        if (nx, ny) not in visited:
            path.append(((nx, ny), rule))
            if water_jug_dfs(nx, ny, goal, visited, path):
                return True
            path.pop()
    
    return False

# Solve
visited = set()
path = []

if water_jug_dfs(0, 0, (0, 4), visited, path):
    print("Water Jug Problem - DFS Solution\n")
    print(f"{'Step':<6} {'State':<15} {'Rule':<30}")
    print("-" * 55)
    print(f"{'0':<6} {str((0,0)):<15} {'Start':<30}")
    
    for i, (state, rule) in enumerate(path, 1):
        print(f"{i:<6} {str(state):<15} {rule:<30}")
    
    print(f"\n✓ Solution found in {len(path)} steps!")
else:
    print("No solution!")
```
