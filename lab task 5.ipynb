# Lab 5 - OpenCV Functions

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create blank image
img = np.zeros((300, 400, 3), dtype=np.uint8)

img[:, :130] = [0, 255, 255]     # Yellow
img[:, 130:260] = [255, 0, 255]  # Purple
img[:, 260:] = [255, 255, 0]     # Cyan

cv2.circle(img, (200, 150), 50, (255, 255, 255), -1)
cv2.putText(img, "AFTAB", (130, 280), 
            cv2.FONT_HERSHEY_SIMPLEX, 
            1, (0, 0, 0), 2)
cv2.imwrite('modified_sample.jpg', img)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
resized = cv2.resize(img, (180, 120))
print("New size:", resized.shape)

plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(gray.shape)
plt.imshow(gray, cmap='gray')
plt.show()
```

```python
canvas = np.ones((300, 300, 3), dtype=np.uint8) * 255

cv2.line(canvas, (50, 50), (250, 50), (255, 0, 0), 2)
cv2.rectangle(canvas, (50, 100), (150, 180), (0, 255, 0), 2)
cv2.circle(canvas, (200, 140), 40, (0, 0, 255), 2)
cv2.putText(canvas, 'OpenCV', (80, 250), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
blurred = cv2.GaussianBlur(img, (5, 5), 0)
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
edges = cv2.Canny(img, 100, 200)
plt.imshow(edges, cmap='gray')
plt.show()
```

```python
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
plt.imshow(thresh, cmap='gray')
plt.show()
```
